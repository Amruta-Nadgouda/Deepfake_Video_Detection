This notebook shows the training and validation accuracies comparison between the models as well as precision recall comparision of all the models

```python
import numpy as np 
import matplotlib.pyplot as plt

models = ['ResNet-50+LSTM','ResNet-152+LSTM','ResNext-50+LSTM','ResNext-101+LSTM', 'Meso-4', 'MesoInception-4', 'Meso-4+MesoInception-4']

# create data
x = np.arange(len(models))
train = [88.29, 88.75, 87.72, 89.37, 81.81, 83.05, 73]
val = [95.22, 95.36, 95.59, 93.96, 80.67, 82.06, 74]
width = 0.3

plt.figure(figsize=(14, 6))
# plot data in grouped manner of bar type
plt.bar(x-0.3, train, width, color='#69b3a2')
plt.bar(x, val, width)
# plt.bar(x+0.3, acc, width, color='#9E39E5')

plt.xticks(x, models, rotation = 25)
plt.xlabel("Model",fontsize=12, fontweight='bold')
plt.ylabel("Accuracy in %", fontsize=12, fontweight='bold')
plt.title("Training and Validation Accuracy of the Models", fontsize=14.0, fontweight='bold')
plt.legend(["train", "val"])
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

models = ['ResNet-50+LSTM','ResNet-152+LSTM','ResNext-50+LSTM','ResNext-101+LSTM', 'Meso-4', 'MesoInception-4', 'Meso-4+MesoInception-4']

# create data
x = np.arange(len(models))
prec = [0.65, 0.66, 0.68, 0.7, 0.55, 0.55, 0.57]
rec = [0.3, 0.35, 0.24, 0.23, 0.57, 0.62, 0.6]
width = 0.3

plt.figure(figsize=(12, 6))
plt.bar(x-0.3, prec, width, color='#823CE7')
plt.bar(x, rec, width, color='#DF60B5')

plt.xticks(x, models, rotation = 25)
plt.title("Precision and Recall Values of the Models", fontsize=14, fontweight='bold')
plt.xlabel("Model",fontsize=12, fontweight='bold')
plt.ylabel("Precision and Recall",fontsize=12, fontweight='bold')
plt.legend(["precision", "recall"])
plt.show()
```
